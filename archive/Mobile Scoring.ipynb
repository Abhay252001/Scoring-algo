```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("test.csv")
df.head()
```

```python
#Set beneficial and Non beneficial criteria to all
beneficial = df[['battery_power','blue','clock_speed','dual_sim','int_memory','n_cores','pc','px_height','px_width'
                 ,'ram','sc_h','three_g','touch_screen','wifi']]
print(beneficial.head())
```

```python
Non_beneficial = df[['mobile_wt']]
print(Non_beneficial.head())
```

```python
for i in beneficial:
    column = beneficial[i]
    max=column.max()
#     print(max)
    beneficial[i] = (beneficial[i] / max)

beneficial
        
```

```python
for i in Non_beneficial:
    column = Non_beneficial[i]
    min=column.min()
#     print(max)
    Non_beneficial[i] = (min / Non_beneficial[i])

Non_beneficial
```

```python
frames = [beneficial, Non_beneficial]
df2 = pd.concat([beneficial, Non_beneficial], axis=1)
df2
```
